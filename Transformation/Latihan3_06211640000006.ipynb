```python
import pandas as pd
import numpy as np
```

```python
pcademo=pd.read_csv('pcademo.csv')
pcademo.head()
```

```python
x=pcademo.iloc[0:51,1:20]
x.tail()
```

```python
#normalization
from sklearn.preprocessing import normalize
xnorm=normalize(x,norm='l2') #default norm l2, other are l1 and max
xnorm=pd.DataFrame(xnorm)
xnorm.tail()
```

```python
#minmax
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(-5,1))
scaler.fit(x)
x_scaled=scaler.transform(x)
x_scaled=pd.DataFrame(x_scaled)
x_scaled.tail()
```

```python
#3sigmoid
xsigmoid=(1-2.718281828**(-xnorm))/(1+2.718281828**(-xnorm))
xsigmoid=pd.DataFrame(xsigmoid)
xsigmoid.tail()
```

```python
#decimal scalling
xdecimal=x/(10**3)
xdecimal=pd.DataFrame(xdecimal)
xdecimal.tail()
```

```python
#softmax
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x), axis=0)
xsoft=softmax(xnorm)
xsoft=pd.DataFrame(xsoft)
print(xsoft)
```

```python
#feature se;ection
from sklearn.feature_selection import SelectKBest, chi2, f_regression
Xx=x.iloc[0:51,2:20]
y=x.iloc[0:51,1]
selector=SelectKBest(f_regression, k="all").fit(Xx,y)
X_new=selector.transform(Xx)
scores=selector.scores_
scores
```

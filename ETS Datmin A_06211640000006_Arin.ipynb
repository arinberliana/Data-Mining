```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
data=pd.read_csv('online_retail.csv')
data.head()
```

```python
np.sum(data.isnull())
```

```python
for i, j in data[data['Description'].isnull()].iterrows():
    data['Description'][i]="Code-"+str(j['StockCode'])
```

```python
np.sum(data.isnull())
```

```python
desc=data.groupby('InvoiceNo')
online=[]
for i,j in desc:
    online.append(list(j['Description'].map(str)))
```

```python
online=pd.DataFrame(online)
online
```

```python
association_rules=apriori(online,min_support=0.0045,
                         min_confidence=0.2,
                         min_lift=3)
association_result=list(association_rules)
```

```python
association_result[0]
```

```python
for item in association_result:
    pair = item[0]
    items = [y for y in pair]
    print("rule: " + items[0] + "->" + items[1])
    print("support: " + str(item[1])) 
    print("confidence: " + str(item[2][0][2]))
    print("lift: " + str(item[2][0][3]))
    print("=====================================")
```
